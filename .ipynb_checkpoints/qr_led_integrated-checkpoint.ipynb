## 1. Import Libraries and Initialize Robot

```python
import cv2
import ipywidgets as widgets
from IPython.display import display, clear_output
import time
from collections import deque
from tiki.mini import TikiMini

# ë¡œë´‡ ì´ˆê¸°í™”
tiki = TikiMini()
print("TikiMini ì´ˆê¸°í™” ì™„ë£Œ!")
```

## 2. LED Control Functions

```python
def clear_leds():
    """ëª¨ë“  LED ë„ê¸°"""
    for i in range(16):
        tiki.set_led(0, i, 0, 0, 0)

def draw_shape(shape, color):
    """
    LED íŒ¨í„´ ì¶œë ¥
    shape: 'X', 'O', '#'
    color: (R, G, B) íŠœí”Œ (0-50)
    """
    clear_leds()
    r, g, b = color
    
    if shape == 'X':
        indices = [15, 0, 9, 6, 10, 5, 12, 3]
    elif shape == 'O':
        indices = [15, 8, 7, 0, 12, 11, 4, 3, 14, 13, 1, 2]
    elif shape == '#':
        indices = list(set([8, 9, 10, 11, 7, 6, 5, 4, 14, 9, 6, 1, 13, 10, 5, 2]))
    else:
        print(f"Unknown shape: {shape}")
        return
    
    for i in indices:
        tiki.set_led(0, i, r, g, b)

print("LED ì œì–´ í•¨ìˆ˜ ë¡œë“œ ì™„ë£Œ!")
```

## 3. QR Code Mapping Configuration

```python
# QR ì½”ë“œì™€ LED íŒ¨í„´ ë§¤í•‘
qr_mapping = {
    'ID_O_R': ('O', (50, 0, 0)),   # ë¹¨ê°„ìƒ‰ O
    'ID_O_G': ('O', (0, 50, 0)),   # ì´ˆë¡ìƒ‰ O
    'ID_O_B': ('O', (0, 0, 50)),   # íŒŒë€ìƒ‰ O
    'ID_X_R': ('X', (50, 0, 0)),   # ë¹¨ê°„ìƒ‰ X
    'ID_X_G': ('X', (0, 50, 0)),   # ì´ˆë¡ìƒ‰ X
    'ID_X_B': ('X', (0, 0, 50)),   # íŒŒë€ìƒ‰ X
    'ID_#_R': ('#', (50, 0, 0)),   # ë¹¨ê°„ìƒ‰ #
    'ID_#_G': ('#', (0, 50, 0)),   # ì´ˆë¡ìƒ‰ #
    'ID_#_B': ('#', (0, 0, 50)),   # íŒŒë€ìƒ‰ #
}

print("QR ë§¤í•‘ ì„¤ì • ì™„ë£Œ!")
print(f"ë“±ë¡ëœ QR ì½”ë“œ: {list(qr_mapping.keys())}")
```

## 4. QR Code Recognition with LED Output

```python
# CSI ì¹´ë©”ë¼ íŒŒì´í”„ë¼ì¸
pipeline = (
    "nvarguscamerasrc ! "
    "video/x-raw(memory:NVMM), width=640, height=480, format=NV12, framerate=30/1 ! "
    "nvvidconv ! video/x-raw, format=BGRx ! "
    "videoconvert ! video/x-raw, format=BGR ! "
    "appsink max-buffers=1 drop=true"  # í•µì‹¬: ë²„í¼ ìµœì†Œí™”
)
cap = cv2.VideoCapture(pipeline, cv2.CAP_GSTREAMER)

# ì¶”ê°€ ìµœì í™”
cap.set(cv2.CAP_PROP_BUFFERSIZE, 1)  # OpenCV ë²„í¼ë„ ìµœì†Œí™”

if not cap.isOpened():
    print("âŒ Camera Open Error")
else:
    print("âœ… Camera Open Success")
    
    # ë¹„ë””ì˜¤ ìœ„ì ¯ ìƒì„±
    video_widget = widgets.Image(format='jpeg', layout=widgets.Layout(width='640px', height='480px'))
    display(video_widget)
    
    def frame_to_bytes(frame):
        # JPEG í’ˆì§ˆ ë” ë‚®ì¶¤ (40 â†’ 30)
        _, buf = cv2.imencode('.jpg', frame, [int(cv2.IMWRITE_JPEG_QUALITY), 30])
        return buf.tobytes()
    
    def get_side_rois(frame):
        """ì¢Œìš° ì¸¡ë©´ ì˜ì—­ ì¶”ì¶œ"""
        h, w = frame.shape[:2]
        # ì™¼ìª½ 40%, ì˜¤ë¥¸ìª½ 40% ì˜ì—­ (ë” ë„“ê²Œ)
        left_roi = frame[:, 0:int(w*0.4)]
        right_roi = frame[:, int(w*0.6):w]
        return [left_roi, right_roi]
    
    def preprocess_simple(roi):
        """ìµœì†Œí•œì˜ ì „ì²˜ë¦¬ (ì†ë„ ìš°ì„ )"""
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        
        # CLAHEë§Œ ì ìš© (ê°€ì¥ íš¨ê³¼ì )
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        enhanced = clahe.apply(gray)
        
        return [gray, enhanced]  # 2ê°€ì§€ë§Œ ë°˜í™˜
    
    import numpy as np
    qr_detector = cv2.QRCodeDetector()
    log_buffer = deque(maxlen=5)
    
    frame_count = 0
    qr_history = {}
    debounce_time = 5
    
    print("\nğŸš€ ì¸¡ë©´ QR ì½”ë“œ ì¸ì‹ ì‹œì‘! (Kernel â†’ Interruptë¡œ ì¤‘ì§€)\n")
    
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            
            frame_count += 1
            frame = cv2.flip(frame, -1)
            
            data, bbox = None, None
            found_side = ""
            
            # === 2í”„ë ˆì„ë§ˆë‹¤ë§Œ QR ê²€ì¶œ (ì†ë„ í–¥ìƒ) ===
            if frame_count % 2 == 0:
                side_rois = get_side_rois(frame)
                side_names = ["LEFT", "RIGHT"]
                
                for side_idx, roi in enumerate(side_rois):
                    if data:
                        break
                    
                    # ì›ë³¸ + ìµœì†Œ ì „ì²˜ë¦¬ë§Œ
                    processed = preprocess_simple(roi)
                    
                    for proc_img in [roi] + processed:
                        data, bbox_roi, _ = qr_detector.detectAndDecode(proc_img)
                        if data:
                            found_side = side_names[side_idx]
                            if bbox_roi is not None and side_idx == 1:
                                bbox_roi[:, :, 0] += int(frame.shape[1] * 0.6)
                            bbox = bbox_roi
                            break
                    
                    # íšŒì „ ì‹œë„ ì¤„ì´ê¸° (Â±10ë„ë§Œ)
                    if not data:
                        for angle in [-10, 10]:
                            h, w = roi.shape[:2]
                            center = (w // 2, h // 2)
                            matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
                            rotated = cv2.warpAffine(roi, matrix, (w, h))
                            data, bbox_roi, _ = qr_detector.detectAndDecode(rotated)
                            if data:
                                found_side = f"{side_names[side_idx]}({angle}Â°)"
                                bbox = bbox_roi
                                break
            
            if data:
                current_time = time.time()
                current_time_str = time.strftime("%H:%M:%S")
                
                can_process = False
                
                if data not in qr_history:
                    can_process = True
                elif (current_time - qr_history[data]) >= debounce_time:
                    can_process = True
                
                if can_process:
                    qr_history[data] = current_time
                    
                    if data in qr_mapping:
                        shape, color = qr_mapping[data]
                        draw_shape(shape, color)
                        log_buffer.append(f"âœ… [{current_time_str}] {data} â†’ {shape} ëª¨ì–‘, RGB{color} ({found_side})")
                    else:
                        log_buffer.append(f"âš ï¸ [{current_time_str}] {data} (ë§¤í•‘ ì—†ìŒ, {found_side})")
                    
                    # ë¡œê·¸ ì¶œë ¥ ìµœì†Œí™”
                    clear_output(wait=True)
                    display(video_widget)
                    print("\nğŸ“‹ QR ì¸ì‹ ë¡œê·¸:")
                    for line in log_buffer:
                        print(line)
                
                # QR ì‹œê°í™” (ê°„ì†Œí™”)
                if bbox is not None:
                    pts = bbox.astype(int).reshape(-1, 2)
                    for i in range(len(pts)):
                        cv2.line(frame, tuple(pts[i]), tuple(pts[(i + 1) % len(pts)]), (0, 255, 0), 2)
            
            # ROI í‘œì‹œ ì œê±° (ì†ë„ í–¥ìƒ)
            # í•„ìš”ì‹œ ì£¼ì„ í•´ì œ
            # h, w = frame.shape[:2]
            # cv2.line(frame, (int(w*0.4), 0), (int(w*0.4), h), (255, 0, 0), 1)
            # cv2.line(frame, (int(w*0.6), 0), (int(w*0.6), h), (255, 0, 0), 1)
            
            video_widget.value = frame_to_bytes(frame)
    
    except KeyboardInterrupt:
        print("\nâ¹ï¸ ì‚¬ìš©ì ì¤‘ì§€")
    
    finally:
        cap.release()
        clear_leds()
        print("\nâœ… ì¹´ë©”ë¼ ì¢…ë£Œ ë° LED OFF")
```

```python
# LED ë„ê¸°
clear_leds()
print("LED OFF")
```
