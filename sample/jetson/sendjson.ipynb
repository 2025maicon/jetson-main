```python
# jetson to 대시보드 임무코드.json 전송 예제

import json
import requests


data = {
    "mission_code": "####",
    "fire_buildings": ["sector2", "sector3"],
    "points": ["alpha","bravo","charlie"],
    "detection": {
        "Alpha": [
            {"type": "hazmat", "count": 1},
            {"type": "enemy", "count": 2},
            {"type": "tank", "count": 1},
            {"type": "mortar", "count": 2},
            {"type": "box", "count": 1},
            {"type": "box2", "count": 2}
        ],
        "Bravo": [
            {"type": "hazmat", "count": 1},
            {"type": "enemy", "count": 3},
            {"type": "tank", "count": 0}
        ],
        "Charlie": [
            {"type": "tank", "count": 1},
            {"type": "mortar", "count": 1},
            {"type": "ammo", "count": 2},
            {"type": "ammo2", "count": 3}
        ]
    }
}


json_filename = "####.json"   # 임무코드.json


def send_dashboard(json_obj, filename):
    try:
        json_content = json.dumps(json_obj, indent=2, ensure_ascii=False)

        files = {
            'file': (filename, json_content, 'application/json')
        }

        print(f"서버로 전송 중: {filename} ...")

        response = requests.post(
            'http://58.229.150.23:5000/dashboard_json',
            files=files,
            timeout=10
        )

        print("서버 응답:", response.text)
    
    except Exception as e:
        print("서버 전송 실패:", e)


send_dashboard(data, json_filename)
```
