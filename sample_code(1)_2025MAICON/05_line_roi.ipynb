```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('road_line.png')

# ROI 영역 설정 (예: 하단 1/3, 좌우 중앙 1/2)
h, w = img.shape[:2]  # get image height (h) and width (w)
roi_y1 = int(h * 0.60) # start ROI at 60% from top (bottom area only)
roi_y2 = h
roi_x1 = int(w * 0.25) # start X ROI at 25% from left (crop left side)
roi_x2 = int(w * 0.75) # end X ROI at 75% from left (crop right side)

roi = img[roi_y1:roi_y2, roi_x1:roi_x2]

gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)

# threshold
_, mask = cv2.threshold(gray, 180, 255, cv2.THRESH_BINARY)

result = cv2.bitwise_and(roi, roi, mask=mask)

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```
